# Data Cleaning

```python
import re #regular expression
import string

def clean_Text(Text):
    '''Make Text lowercase, remove Text in square brackets, remove punctuation and remove words containing numbers.'''
    Text = Text.lower()
    Text = re.sub('\[.*?\]', '', Text)
    Text = re.sub('[%s]' % re.escape(string.punctuation), '', Text)
    Text = re.sub('\w*\d\w*', '', Text)
    Text = re.sub("[0-9" "]+"," ",Text)
    Text = re.sub('[‘’“”…]', '', Text)
    return Text

clean = lambda x: clean_Text(x)
```

```python
data['Text'] = data.Text.apply(clean)
data.Text
```

```python
# Function to compute N-grams
def compute_ngrams(tokens, n):
    return list(ngrams(tokens, n))
```

```python
# Compute unigrams
unigrams = compute_ngrams(data1, 1)
unigrams
```

```python
bigrams = compute_ngrams(data1, 2)
bigramsrams
```

```python
trigrams = compute_ngrams(data1, 3)
trigrams
```

```python
from collections import Counter
```

```python
# Print the most common N-grams
print("Top 10 most common unigrams:")
print(Counter(unigrams).most_common(10))
print("\nTop 10 most common bigrams:")
print(Counter(bigrams).most_common(10))
print("\nTop 10 most common trigrams:")
print(Counter(trigrams).most_common(10))
```

```python
import matplotlib.pyplot as plt

# Function to plot N-grams
def plot_ngrams(ngrams_list, title):
    ngram_counts = Counter(ngrams_list).most_common(10)
    labels, counts = zip(*ngram_counts)
    # Convert labels to strings
    labels = [' '.join(map(str, label)) for label in labels]
    plt.figure(figsize=(10, 6))
    plt.barh(labels, counts)
    plt.xlabel('Frequency', fontsize=15)
    plt.ylabel('N-gram', fontsize=15)
    plt.title(title, fontsize=15) 
    plt.gca().invert_yaxis()  # Invert y-axis to display most common at the top
    plt.show()
```

```python
# Plot top unigrams
plot_ngrams(unigrams, 'Top 10 most common unigrams')
```

```python
# Plot top bigrams
plot_ngrams(bigrams, 'Top 10 most common bigrams')
```

```python
# Plot top trigrams
plot_ngrams(trigrams, 'Top 10 most common trigrams')
```

```python
import matplotlib.pyplot as plt

# Define the function to get top uni-grams
def get_top_n_words(corpus, n=None):
    vec1 = CountVectorizer(max_features=2000).fit(corpus)
    bag_of_words = vec1.transform(corpus)
    sum_words = bag_of_words.sum(axis=0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec1.vocabulary_.items()]
    words_freq = sorted(words_freq, key=lambda x: x[1], reverse=True)
    return words_freq[:n]

# Extract top 200 uni-grams
top_words = get_top_n_words(data["Text"], n=200)
top_df = pd.DataFrame(top_words)
top_df.columns = ["Uni-gram", "Freq"]

# Display the first 50 rows
top_df_50 = top_df.head(50)

# Plot the top 50 uni-grams
plt.figure(figsize=(10, 8))
plt.barh(top_df_50["Uni-gram"], top_df_50["Freq"], color='skyblue')
plt.xlabel('Frequency', fontsize=12)
plt.ylabel('Uni-gram', fontsize=12)
plt.title('Top 50 Uni-grams', fontsize=14)
plt.gca().invert_yaxis()  # Invert y-axis to display most common at the top
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define the function to get top bi-grams
def get_top_n2_words(corpus, n=None):
    vec1 = CountVectorizer(ngram_range=(2, 2), max_features=2000).fit(corpus)
    bag_of_words = vec1.transform(corpus)
    sum_words = bag_of_words.sum(axis=0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec1.vocabulary_.items()]
    words_freq = sorted(words_freq, key=lambda x: x[1], reverse=True)
    return words_freq[:n]

# Extract top 200 bi-grams
top2_words = get_top_n2_words(data["Text"], n=200)
top2_df = pd.DataFrame(top2_words)
top2_df.columns = ["Bi-gram", "Freq"]

# Display the first 50 rows
top2_df_50 = top2_df.head(50)

# Plot the top 50 bi-grams
plt.figure(figsize=(10, 8))
plt.barh(top2_df_50["Bi-gram"], top2_df_50["Freq"], color='skyblue')
plt.xlabel('Frequency', fontsize=12)
plt.ylabel('Bi-gram', fontsize=12)
plt.title('Top 50 Bi-grams', fontsize=14)
plt.gca().invert_yaxis()  # Invert y-axis to display most common at the top
plt.show()
```

```python
#Tri-gram
def get_top_n3_words(corpus, n=None):
    vec1 = CountVectorizer(ngram_range=(3,3),
           max_features=2000).fit(corpus)
    bag_of_words = vec1.transform(corpus)
    sum_words = bag_of_words.sum(axis=0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in
                  vec1.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1],
                reverse=True)
    return words_freq[:n]
```
